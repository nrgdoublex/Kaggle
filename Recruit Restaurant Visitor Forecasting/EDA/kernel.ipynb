# 1. Exploring Raw Data

## 1.1 Air Visit Data

```python
# see total visitors by date
air_visitor = pd.read_csv('../data/air_visit_data.csv')
air_visitor['visit_date'] = pd.to_datetime(air_visitor['visit_date'])
air_visitor_total_by_date = air_visitor.groupby(['visit_date']).sum()
air_visitor_total_by_date.plot(title='Total Visitors by Date')
plt.show()
```

There's periodicity in the data, which I suspect is due to weekdays. Also, there's a huge gap before and after 2016-07-01. I don't know the reason but we can take a more deep look at it later.
In the following I would look at weekdays.

```python
# merge air visitors and date info
date_info = pd.read_csv('../data/date_info.csv')
date_info['calendar_date'] = pd.to_datetime(date_info['calendar_date'])
date_info.rename(columns={'calendar_date':'visit_date'},inplace=True)

air_visitor_total_by_date = air_visitor_total_by_date.reset_index()
air_visitor_total_by_date = air_visitor_total_by_date.merge(date_info,on='visit_date',how='left')
```

### Explore Weekdays

```python
# see total visitors each weekday
data = air_visitor_total_by_date.groupby('day_of_week').sum()
data = data.reindex(['Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday'])
data['visitors'].plot(kind='bar')
plt.show()
```

Weekends (Fri. and Sat.) have more visitors than weekdays. Note that Sunday doesn't have much more visitors than other weekdays, so we regard it as weekdays. By this exploration we may create a feature called 'weekend' to indicate if a specific day is weekend or not.

### Explore Holidays

```python
data = air_visitor_total_by_date.groupby('holiday_flg').sum()
data.plot(kind='bar',title='Total Visitors')
plt.show()
data = air_visitor_total_by_date.groupby('holiday_flg').mean()
data.plot(kind='bar',title='Mean Visitors')
plt.show()
```

This seems strange because we expect in holidays there will be more people having free time to go to restaurants. It's necessary to see the definition of 'holidays'.

```python
holidays = pd.read_csv('../data/date_info.csv')
print "Number of Holidays as Saturday: {}".format(len(holidays[(holidays['day_of_week'] == 'Saturday') & (holidays['holiday_flg'] == 1)]))
print "Number of Holidays as Sunday: {}".format(len(holidays[(holidays['day_of_week'] == 'Sunday') & (holidays['holiday_flg'] == 1)]))
print "Total number of holidays: {}".format(len(holidays[holidays['holiday_flg'] == 1]))
```

From the above observation we know 'holiday' here means national holidays or something else, excluding weekends. One thing of interest to know is that how many restaurants are open during holidays. We can explore that to help us get better idea of how holidays affect the number of visitors.

```python
data = air_visitor.merge(date_info,on='visit_date',how='left')
num_total_ids = len(data['air_store_id'].unique())
num_holiday_ids = len(data[data['holiday_flg'] == 1]['air_store_id'].unique())
print "Number of total restaurants: {}".format(num_total_ids)
print "Number of restaurants open in at least one holiday: {}".format(num_holiday_ids)
```

Here we can make sure that the fewer visitors in holidays are due to less number of holidays compared to non-holidays, which means 'holiday' may not be a good feature. But considering many restaurants have the same off-days annually, maybe we can figure out if a restaurant is off at some holidays and it may be off at the same day for the years to come. Let's take one example to observe that.

```python
air_id = 'air_25e9888d30b386df'
data = air_visitor.merge(date_info,on='visit_date',how='left')
data = data[(data['air_store_id'] == air_id) & (data['holiday_flg'] == 1)]
print data.head()
```

We hypothesize that many restaurants are off at certain holidays, and they will be off at the same holidays for the years to come.

### Explore Month

```python
# see total visitors each month
air_visitor['month'] = air_visitor['visit_date'].map(lambda x: x.month)
air_visitor['year'] = air_visitor['visit_date'].map(lambda x: x.year)
air_visitor['year_month'] = air_visitor['visit_date'].map(lambda x: str(x.year)+'/'+str(x.month))
data = air_visitor.groupby(['year','month']).sum()
data['visitors'].plot(kind='bar',title='Total Visitors')
plt.show()
data = air_visitor.groupby(['year','month']).mean()
data['visitors'].plot(kind='bar',title='Mean Visitors')
plt.show()
```

The mean visitors shows little gap while the total visitors has a gap before and after 2016/07. This may imply there is an issue about the number of open restaurants. Let's check this.

```python
air_visitor['month'] = air_visitor['visit_date'].map(lambda x: x.month)
air_visitor['year'] = air_visitor['visit_date'].map(lambda x: x.year)
air_visitor['year_month'] = air_visitor['year'].map(str) + '/' + air_visitor['month'].map(str)
for year_month in air_visitor.year_month.unique():
    year, month = map(lambda x: int(x),year_month.split('/'))
    data = air_visitor[(air_visitor['year'] == year) & (air_visitor['month'] == month)]
    print "Year: {}, Month: {}, Total restaurants: {}".format(year,month,len(data['air_store_id'].unique()))
```

We find out why there's a huge gap of total visitors before and after 2017/07, as we expected. But here comes another issue: if one restaurant didn't appear before 2016/07, how do we know how features like 'holiday' and 'month' influence the visitors?

To end this subsection, we observe the trend from 2016-04-15 to 2016-06-15, which is the same period as the test set.

```python
# check the same period as test data in 2016
air_visitors = pd.read_csv('../data/air_visit_data.csv')
all_air_visitors = air_visitors.groupby('visit_date').sum()
visitors_apr_jun = all_air_visitors.loc['2016-04-15':'2016-06-15','visitors']
visitors_apr_jun.plot()
plt.show()
```

This also shows periodicity but there was also a small fluctuation during golden weeks, which is between 2016-04-29 and 2016-05-05.

## 1.2 Air Reserves

First observe the layout of data

```python
# visualize air reserves
air_reserve = pd.read_csv('../data/air_reserve.csv')
air_reserve['visit_date'] = pd.to_datetime(air_reserve['visit_datetime']).dt.date
air_reserve['visit_time'] = pd.to_datetime(air_reserve['visit_datetime']).dt.time
air_reserve['reserve_date'] = pd.to_datetime(air_reserve['reserve_datetime']).dt.date
air_reserve['reserve_time'] = pd.to_datetime(air_reserve['reserve_datetime']).dt.time
air_reserve.head()
```

Then we plot the trend of visit date and reserve date

```python
air_reserve_all_visitors = air_reserve.groupby('visit_date').sum()
air_reserve_all_visitors.plot(title='Total visitors by visit date')
plt.show()
air_reserve_all_reservers = air_reserve.groupby('reserve_date').sum()
air_reserve_all_reservers.plot(title='Total visitors by reserve date')
plt.show()
```

We cannot see any clear trends from the above two plots. Let's turn to the visit time.

```python
# visualize their visit time
air_reserve_hours = air_reserve.groupby('visit_time')['reserve_visitors'].sum()
air_reserve_hours.plot(kind = 'bar')
plt.show()

# shows most of reservations are at dinner time
```

It is clear the visit times concentrate on the dinner time, which is reasonable. Next we observe the time difference between reserve and visit time.

```python
air_reserve['visit_datetime'] = pd.to_datetime(air_reserve['visit_datetime'])
air_reserve['reserve_datetime'] = pd.to_datetime(air_reserve['reserve_datetime'])
air_reserve['time_diff'] = air_reserve['visit_datetime'].sub(air_reserve['reserve_datetime'])
air_reserve_timediff = air_reserve.groupby('time_diff').sum()['reserve_visitors']
air_reserve_timediff.index = air_reserve_timediff.index.astype('timedelta64[h]')
print air_reserve_timediff.index


segment = air_reserve_timediff.head(200)
segment.plot(kind='bar')
#plt.xticks(np.arange(min(segment), max(segment)+1, 1.0))
plt.show()

# shows the time difference between visit and reserve has 24-hour interval
```

We know the reservation within 3 hours counts most of reservations. Also note that there shows the periodicity of 24 hours. I think that's due to the over seat and visitors postpone their visits to tomorrow or days later.

# Feature Selection

## Weekend/Nonweekend

```python
air_visit = pd.read_csv('../data/air_visit_data.csv')
date_info = pd.read_csv('../data/date_info.csv')
air_weekend = air_visit.merge(date_info,left_on='visit_date',right_on='calendar_date',how='left')
air_weekend = air_weekend.drop('calendar_date',1)
air_weekend['visit_date'] = pd.to_datetime(air_weekend['visit_date'])
air_weekend['visitors_log1p'] = air_weekend['visitors'].map(lambda x: np.log1p(x))
air_weekend['weekend'] = air_weekend['day_of_week'].map(lambda x: int(x == 'Friday' or x == 'Saturday'))
sns.boxplot(data=air_weekend,x='weekend',y='visitors_log1p')
plt.show()
```

From the above we expect 'weekend' to be an somewhat effective feature.

## year_month

```python
air_visit['visit_date'] = pd.to_datetime(air_visit['visit_date'])
air_yearmonth = air_visit.copy(deep=True)
air_yearmonth['year_month'] = air_yearmonth['visit_date'].map(lambda x: str(x.year) + '_' + str(x.month))
air_yearmonth['visitors_log1p'] = air_yearmonth['visitors'].map(lambda x: np.log1p(x))
sns.boxplot(data=air_yearmonth,x='year_month',y='visitors_log1p')
plt.show()
```

There's no significant difference between months. Maybe we can ignore this feature?

## location

```python
air_store = pd.read_csv('../data/air_store_info.csv')
air_store['prefecture'] = air_store['air_area_name'].map(lambda x: x.split()[0].decode('utf-8'))
air_loc = air_visit.merge(air_store,on='air_store_id',how='left')
air_loc['visitors_log1p'] = air_loc['visitors'].map(lambda x: np.log1p(x))
g = sns.boxplot(data=air_loc,x='prefecture',y='visitors_log1p')
plt.xticks(rotation=45)
plt.show()
```

There are some differences bewteen every prefecture. So maybe this can be a feature.

```python
g = sns.boxplot(data=air_loc,x='air_genre_name',y='visitors_log1p')
plt.xticks(rotation=45)
plt.show()
```

Considering so much variation in the above boxplot, we definitely count in genre as one of features.
